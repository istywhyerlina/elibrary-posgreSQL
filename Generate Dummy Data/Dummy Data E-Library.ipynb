```python
from faker import Faker
import random
import re

n_user=500
fake = Faker('id_ID')

def generate_dummy_user(n):
    """
    A function to generate user data by Faker.

    Args
    n (int) : amount of users to be generated

    Return
    user_data (list)
    """
    user_data = []

    for i in range(1,n+1):
        pn=fake.phone_number()
        new_pn=re.sub(r'[^\w]', '', pn)
        created_date = fake.date_time_between_dates(datetime_start =datetime(2023, 1, 1, 0, 0, 0), datetime_end =datetime(2023, 10, 31, 23, 59, 59)).strftime("%Y-%m-%d %H:%M:%S")
        user_data.append({
            "user_id": i,
            "name": fake.name(),
            "email":fake.email(),
            "phone_number": new_pn,
            "date_created": created_date
        })

    return user_data




#define column for user table
user_cols = ['user_id', 'name','email' ,'phone_number','date_created']
#generate rows for user table
user_rows = generate_dummy_user(n_user)
user_df=pd.DataFrame(user_rows, columns=user_cols)
user_df
```

```python
#generate book_availability
#generate current queue & its requests
n=400

book_availability_data = []
queue_data=[]
requests_data=[]
k=1   
for i in range(1,n+1):
        previous=None
        library_id=random.choice(libraries['library_id'])
        book_id=random.choice(books['book_id'])
        quantity=random.randint(0, 10)
        quantity_available=random.randint(0, quantity)
        queue=random.randint(0, 10) if quantity_available==0 else 0
        book_availability_data.append({
            "ID": i,
            "library_id":library_id ,
            "book_id":book_id,
            "quantity_owned":  quantity,
            "quantity_available": quantity_available,
            "queue": queue
        })
        for j in range(queue):
            user_chosen = random.choice(user_df['user_id'])
            request_date=fake.date_time_between_dates(datetime_start =datetime(2024, 1, 1, 23, 59, 59), datetime_end =datetime(2024, 1, 31, 23, 59, 59)).strftime("%Y-%m-%d %H:%M:%S")
            request_id=3000+k
            queue_id=3000+k
            requests_data.append({
            "request_id": request_id,
            "user_id": user_chosen,
            "library_id": library_id,
            "book_id":book_id,
            "request_date":  request_date,
            "request_status": 'Borrow',
            "previous_request_to_excecute":previous
            })

            queue_data.append({
            "queue_id": queue_id,
            "request_id": request_id,
            "is_queue_active":True
            })
            k+=1
            
            




#define column for book_availability table
book_availability_cols = ['ID', 'library_id','book_id' ,'quantity_owned', 'quantity_available','queue']
#generate rows for book_availability table
book_availability_df=pd.DataFrame(book_availability_data, columns=book_availability_cols)
book_availability_df
```

```python
#generate data for returned book

n=300
borrow_data = []
return_data = []
hold_data=[]

    
for i in range(1,n+1):
        request_id=500+i
        previous_request=i
        user_chosen = random.choice(user_df['user_id'])
        date_min= datetime.strptime( user_df['date_created'][user_chosen-1], "%Y-%m-%d %H:%M:%S")
        borrow_date=fake.date_time_between_dates(datetime_start =date_min, datetime_end =datetime(2023, 10, 31, 23, 59, 59)).strftime("%Y-%m-%d %H:%M:%S")
        previous_request_date=fake.date_time_between_dates(datetime_start =date_min, datetime_end =datetime.strptime(borrow_date, "%Y-%m-%d %H:%M:%S")).strftime("%Y-%m-%d %H:%M:%S")        
        return_date_start=datetime.strptime(borrow_date, "%Y-%m-%d %H:%M:%S")
        diff_day_return=random.randint(1,14)
        is_automatic=True if diff_day_return==14 else False 
        return_date_end=datetime.strptime(borrow_date, "%Y-%m-%d %H:%M:%S")+timedelta(days=diff_day_return)
        return_date=fake.date_time_between_dates(datetime_start =return_date_start, datetime_end =return_date_end)
        start_hold_date_start=datetime.strptime(borrow_date, "%Y-%m-%d %H:%M:%S")-timedelta(days=7)
        start_hold_date=fake.date_time_between_dates(datetime_start =start_hold_date_start, datetime_end =datetime.strptime(borrow_date, "%Y-%m-%d %H:%M:%S")).strftime("%Y-%m-%d %H:%M:%S")
        library=random.choice(book_availability_df['library_id'])
        book=random.choice(book_availability_df['book_id'])
        previous1=None
        previous2=int(previous_request)
        borrow_data.append({
            "borrow_id": i,
            "request_id": request_id,
            "hold_id": i,
            "borrow_date":borrow_date,
            "return_id":  i,
        })
        return_data.append({
            "return_id": i,
            "return_date": return_date,
            "is_automatic": is_automatic,

        })
        requests_data.append({
            "request_id": request_id,
            "user_id": user_chosen,
            "library_id": library,
            "book_id":book,
            "request_date":  borrow_date,
            "request_status": 'Borrow After Hold',
            "previous_request_to_excecute":previous2
        })
        requests_data.append({
            "request_id": previous_request,
            "user_id": user_chosen,
            "library_id": library,
            "book_id":book,
            "request_date":  previous_request_date,
            "request_status": 'Borrow',
            "previous_request_to_excecute":previous1
        })

        hold_data.append({
            "hold_id": i,
            "queue_id": i,
            "request_id_after_hold": request_id,
            "start_hold":start_hold_date,
            "finished_hold":  borrow_date,
            "hold_status": 'Borrowed',
        })
        queue_data.append({
            "queue_id": i,
            "request_id": previous_request,
            "is_queue_active":False
        })
        












```

```python
#generate data current borrow
k=1
for i in range (book_availability_df.shape[0]):
    borrowed=(book_availability_df['quantity_owned'][i] - book_availability_df['quantity_available'])[i]
    library_id=book_availability_df['library_id'][i]
    book_id=book_availability_df['book_id'][i]
    for j in range(borrowed):
        borrow_id=300+k
        request_id=1000+k
        user_chosen = random.choice(user_df['user_id'])
        date_requested_borrowed=fake.date_time_between_dates(datetime_start =datetime(2023, 12, 17, 23, 59, 59), datetime_end =datetime(2023, 12, 31, 23, 59, 59)).strftime("%Y-%m-%d %H:%M:%S")
        requests_data.append({
            "request_id": request_id,
            "user_id": user_chosen,
            "library_id": library_id,
            "book_id":book_id,
            "request_date":  date_requested_borrowed,
            "request_status": 'Borrow'
        })
        borrow_data.append({
            "borrow_id": borrow_id,
            "request_id": request_id,
            "hold_id": None,
            "borrow_date":date_requested_borrowed,
            "return_id":  None,
        })
        k+=1
        


```
